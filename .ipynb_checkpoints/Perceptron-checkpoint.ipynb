```python
import numpy as np
import os
import copy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class Perceptron(object):
    def __init__(self, epochs=10, eta=None):
        self.w = None
        self.epochs = epochs
        self.eta = eta
    
    def init_params(self, n_features):
        self.w = np.random.random(size=(n_features + 1, 1))
        
    def fit(self, x, y):
        """
        :parameter x:m * n
        :parameter y:m * 1
        """
        if self.eta is None:
            self.eta = max(1e-2, 1.0 /np.sqrt(x.shape[0]))
        y = y.reshape(-1, 1)
        y[y == 0] = -1
        n_samples, n_features = x.shape
        self.init_params(n_features)
        x = np.c_[x, np.ones(shape=(n_samples,))]
        x_y = np.c_[x, y]
        
        for _ in range(self.epochs):
            error_sum = 0
            np.random.shuffle(x_y)
            for index in range(0, n_samples):
                x_i = x_y[index, :-1]
                y_i = x_y[index, -1:]
                if (x_i.dot(self.w) * y_i)[0] < 0:
                    dw = (-x_i * y_i).reshape(-1, 1)
                    self.w = self.w - self.eta * dw
                    error_sum += 1
            if error_sum ==0:
                break
                
    def get_params(self):
        return self.w
    
    def predict(self, x):
        n_samples = x.shape[0]
        x = np.c_[x, np.ones(shape=(n_samples,))]
        return (x.dot(self.w) > 0).astype(int)
    
    def sigmoid(self, x2):
        x = copy.deepcopy(x2)
        if type(x) is int:
            x = 20.0 if x > 20.0 else x
            x = -100.0 if x < -100.0 else x
        else:
            x[x > 20.0] = 20.0
            x[x < -100.0] = -100.0
        return 1 / (1 + np.exp(-x))

    def predict_proba(self, x):
        n_samples = x.shape[0]
        x = np.c_[x, np.ones(shape=(n_samples,))]
        return self.sigmoid(x.dot(self.w))
    
    def plot_decision_boundary(self, x, y):
        weights = self.get_params()
        w1 = weights[0][0]
        w2 = weights[1][0]
        bias = weights[-1][0]
        x1 = np.arange(np.min(x), np.max(x), 0.1)
        x2 = -w1 / w2 * x1 - bias / w2
        plt.scatter(x[:, 0], x[:, 1], c=y, s=50)
        plt.plot(x1, x2, 'r')
        plt.show()    
```

```python
from sklearn.datasets import make_classification
data,target=make_classification(n_samples=200, n_features=2,n_classes=2,n_informative=1,n_redundant=0,n_repeated=0,n_clusters_per_class=1)
```

```python
plt.scatter(data[:, 0], data[:, 1], c=target,s=50)
```

```python
perceptron = Perceptron()
perceptron.fit(data, target)
#print(perceptron.predict(data))
#print(perceptron.predict_proba(data))
```

```python
perceptron.plot_decision_boundary(data,target)
```
